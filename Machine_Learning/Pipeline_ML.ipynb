```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
```

```python
data_1923=pd.read_csv("lic_clean_2019_2023.csv")
```

```python
data_1923['annee'] = data_1923['annee'].apply(str)
data_1923['code_fed'] = data_1923['code_fed'].apply(str)
data_1923['total_lic'] = data_1923["total_lic"].apply(int)
data_1923['total_f'] = data_1923["total_f"].apply(int)
data_1923['total_h'] = data_1923["total_h"].apply(int)

data_1923 = data_1923.drop(columns=['code_commune', 'nom_commune', 'num_departement', 'region'])
data_1923.info()
```

```python
#Variable nb total de licenciÃ©s
X = data_1923.loc[:, data_1923.columns != "total_lic"]
y=data_1923.loc[:, "total_lic"]
```

```python
#Ratio pour les datasets >1M de lignes : 99/1
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size=0.01,
                                                    random_state=2,
                                                    stratify=y)
```

```python
X_train.info()
```

```python
numeric_features=list(range(3, 16))

categorical_features=list(range(0, 2))


numeric_transformer=StandardScaler()
categorical_transformer = OneHotEncoder()

feature_encoder = ColumnTransformer(
    transformers=[
       ('cat', categorical_transformer, categorical_features),
       ('num', numeric_transformer, numeric_features)
    ]
)

X_train = feature_encoder.fit_transform(X_train)
X_test = feature_encoder.transform(X_test)
```

```python
regressor=LinearRegression
regressor.fit(X_train, y_train)
```
