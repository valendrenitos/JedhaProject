```python
import pandas as pd
```

```python

data_0223 = pd.read_csv("lics_histo(2).csv", delimiter = ";")
```

```python
data_4903 = pd.read_csv("lics_1949_2003.csv", delimiter = ";", encoding="latin-1")
```

```python
data_0223.info()
```

```python
data_0223.head()
```

```python
data_0223["Lics"].replace({float('NaN'): '0'})
data_0223["fede_gp"] = data_0223["fede_gp"].replace({
    "1. Olympique": "olympique",
    '2. Délégataire non-olympique' : "non-olympique",
    '4. Affinitaire' : "affinitaire",
    '5. Para-sport' : 'para-sport',
    "6. Scolaire": "scolaire",
    "0. Toutes fédés": "Toutes fedes"
})
```

```python
data_0223.groupby(
    by=["annee", "fede_gp"],
)["Lics"].agg(['sum']
)
```

```python
pivot = data_0223.pivot_table(
    index="annee",
    columns="fede_gp",
    values="Lics",
    aggfunc='sum'
)
pivot = pivot.reset_index()
print(pivot)
```

```python

data_4903 = data_4903['Année;Clubs;Fédérations Olympiques;Fédérations Non-Olympiques;Fédérations Multisports et Affinitaires;Fédérations Scolaires et Universitaires;Total'].str.split(';', expand=True)
```

```python
data_4903.columns = ['annee', 'club', 'olympique', 'non-olympique', 'affinitaire', 'scolaire', 'total']
data_4903.head()
data_4903['club'] = data_4903["club"].apply(int)
data_4903['olympique'] = data_4903["olympique"].apply(int)
data_4903['non-olympique'] = data_4903["non-olympique"].apply(int)
data_4903['affinitaire'] = data_4903["affinitaire"].apply(int)
data_4903['scolaire'] = data_4903["scolaire"].apply(int)
data_4903['total'] = data_4903["total"].apply(int)

data_4903.info()
```

```python
for col in data_4903.columns:
    if data_4903[col].dtype == "int64":
        data_4903[col] = data_4903[col] * 1000

data_4903.head()
```

```python
pivot.head()
```

```python

data_0223v2=pivot.drop(columns=["Toutes fedes", "para-sport"])
data_0223v2=data_0223v2[[
    "annee", 
    "olympique", 
    "non-olympique",
    "affinitaire", 
    "scolaire"
]]

data_0223v2
```

```python
data_0223v2['annee'] = pd.to_datetime(data_0223v2['annee']).dt.year
data_0223v2['annee'] = data_0223v2['annee'].apply(str)
data_0223v2.head()
```

```python
total=data_0223.groupby(
    by=["annee"],
)["Lics"].agg(['sum'])

total.reset_index(inplace=True)
total['annee'] = pd.to_datetime(total['annee']).dt.year
total['annee'] = total['annee'].apply(str)

print(total)
```

```python
data_4903=data_4903.drop(columns="club")
data_4903['annee'] = data_4903['annee'].apply(str)
```

```python
data_4903=data_4903.drop(data_4903[data_4903['annee'].isin(["2002", "2003"])].index)
```

```python
print(data_4903[data_4903["annee"] == 2002])
```

```python
data_4903.head()
```

```python
data_0223v2 = data_0223v2.merge(total, on="annee", how="left")
print(data_0223v2)
```

```python
data49_23 = pd.concat([data_4903, data_0223v2], ignore_index=True)

data49_23.info()
```
